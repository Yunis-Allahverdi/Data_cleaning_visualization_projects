```python
import pandas as pd


data = pd.read_csv(r"C:\Users\allah\Downloads\archive (4)\Stats survey.csv")
            
data.head(5)
```

```python
cols = data.columns.tolist()
cols
```

```python
cols[1] = "Gender"
cols[2] = "GPA"
cols[3] = "Prev_grade"
cols[4] = "Faculty"
cols[5] = "Prev_gpa"
cols[6] = "Housing_last"
cols[7] = "Spending_monthly"
cols[8] = "Scholarship"
cols[9] = "study_hr_per_week"
cols[10] = "partying_during_week"
cols[11] = "alcoholic_drink_consume"
cols[12] = "classes_miss_for_alcohol"
cols[13] = "exams_failed"
cols[14] = "dating"
cols[15] = "parents_allow_alcohol"
cols[16] = "relationship_with_parents"

data.columns = cols

data.head(5)
```

```python
data.drop('Prev_gpa', axis=1, inplace=True)
data.drop('Housing_last', axis=1, inplace=True)
data.drop('study_hr_per_week', axis=1, inplace=True)
data.drop('alcoholic_drink_consume', axis=1, inplace=True)
```

```python
data.dropna(subset=['Gender'], inplace=True)
data.dropna(subset=['GPA'], inplace=True)
```

```python
data['Prev_grade'] = data['Prev_grade'].fillna("1st Year")
```

```python
data['Faculty'] = data['Faculty'].fillna(data['Faculty'].mode()[0])
data['Spending_monthly'] = data['Spending_monthly'].fillna(data['Spending_monthly'].mode()[0])
data['Scholarship'] = data['Scholarship'].fillna(data['Scholarship'].mode()[0])
data['partying_during_week'] = data['partying_during_week'].fillna(data['partying_during_week'].mode()[0])
data['classes_miss_for_alcohol'] = data['classes_miss_for_alcohol'].fillna(data['classes_miss_for_alcohol'].mode()[0])
data['exams_failed'] = data['exams_failed'].fillna(data['exams_failed'].mode()[0])
data['dating'] = data['dating'].fillna(data['dating'].mode()[0])
data['parents_allow_alcohol'] = data['parents_allow_alcohol'].fillna(data['parents_allow_alcohol'].mode()[0])

data['relationship_with_parents'] = data['relationship_with_parents'].fillna(data['dating'].mode()[0])
```

```python
data['Timestamp'] = pd.to_datetime(data['Timestamp'], errors='coerce')
```

```python
data.drop_duplicates(inplace=True)
```
